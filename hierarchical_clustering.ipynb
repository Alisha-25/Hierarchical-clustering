```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
%matplotlib inline
```

# create a sample data

```python
from sklearn.datasets import make_blobs
```

```python
X, y_true = make_blobs(n_samples=300,
                       centers=4,
                       cluster_std=2.1,
                       random_state=7)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
```

```python
mykmeans = KMeans(n_clusters = 4,random_state=7)
mykmeans.fit(X)
```

```python
y_kmeans = mykmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = mykmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

```python
# run kmeans for 5 cluster
from sklearn.cluster import KMeans
mykmeans = KMeans(n_clusters = 5,random_state=7)
mykmeans.fit(X)
```

```python
y_kmeans = mykmeans.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = mykmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```

# Hierarchical Clustering

### draw dendrogram

```python
import matplotlib.pyplot as plt

import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10, 7))
plt.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(X, method='ward'))
plt.axhline(y=140, color='r', linestyle='--')
```

## Agglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering( n_clusters= 3 , affinity='euclidean', linkage='ward')
cluster_labels = cluster.fit_predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=cluster_labels, cmap='rainbow')
```
